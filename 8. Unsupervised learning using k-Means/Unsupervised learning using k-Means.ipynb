## Unsupervised learning using k-Means.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import random as rd
from sklearn.cluster import KMeans
```

```python
# Dataset 
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
x = df.iloc[:,[3,4]].values
print(x.shape)
```

```python
kmeans5  = KMeans(n_clusters=5)
kmeans5.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
print("Centroids :\n ",kmeans5.cluster_centers_)
```

```python
print("Predicted labels",kmeans5.labels_)
```

```python
plt.scatter(x[:,0],x[:,1], c=kmeans5.labels_, cmap='rainbow')
plt.scatter(kmeans5.cluster_centers_[:,0],kmeans5.cluster_centers_[:,1], color='Black')
```

```python
plt.scatter(x[:,0],x[:,1], c = kmeans5.labels_, cmap='rainbow')
```
