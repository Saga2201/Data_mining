### Objective

Apriori algorithm to find out which items are commonly sold together <br>

#### Application
So that store owner can 
<li> take action to place the related items together 
<li> advertise them together in order to have increased profit

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyfpgrowth as pf
```

```python
df = pd.read_csv('store_data.csv', header=None)
```

#### Data preparation

Transform dataset to be in the form of a list of lists-<br>
<li> Currently the available data is in the form of a pandas dataframe. 
<li> The whole dataset should be a big list and each transaction in the dataset is an inner list within the outer big list. 

```python
records = []
for sublst in df.values.tolist():
    clean_sublst = [item for item in sublst if item is not np.nan]
    records.append(clean_sublst)
```

```python
patterns = pf.find_frequent_patterns(records,2) ## support_threshold = 2
```
