```python
# Now we are going to store each and every record(row) into list so it'll make list of list
# record will contain each transaction list

records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
records
```

```python
# now we are going to take min_support from user for that we have to take items which is least purchased in particular day
# for example item X is purchased 6 times per day so for whole week we are going to multiply 6 * 7 = 42
# so 42/8998(number of records in dataset) = 0.0046 become our min_support.

association_rules = apriori(records, min_support=0.0046, min_confidence=0.20, min_lift=2, min_length=2)
```

```python
association_rules
```

```python
association_results = list(association_rules)
```

```python
print(association_results)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
for item in association_results:
    pair = item[0] 
    items = [i for i in pair]
    #print(items[0])
    print("Rule: " + items[0] + " -> " + items[1] + " --->")

    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
